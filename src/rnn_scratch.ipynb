```python
import tensorflow as tf
import numpy as np
```

```python
inputs = tf.placeholder(tf.float32, shape = [None, 81, 128])
```

```python
forward_cell = tf.nn.rnn_cell.GRUCell(64)
backward_cell = tf.nn.rnn_cell.GRUCell(64)

outputs, _ = tf.nn.bidirectional_dynamic_rnn(
    forward_cell,
    backward_cell,
    inputs,
    dtype = tf.float32)
```

```python
sess = tf.InteractiveSession()
```

```python
sess.run(tf.global_variables_initializer())
```

```python
sess.run(outputs, feed_dict = { inputs: np.random.rand(10, 81, 128) })
```
